```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score
```

## Dataset 1(Immunotherapy)

```python
df=pd.read_excel('Immunotherapy.xlsx')
df.head()
```

```python
df.columns
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df.Result_of_Treatment=df.Result_of_Treatment.astype(str)
```

```python
part_val = df.Result_of_Treatment.value_counts()
```

```python
df.isnull().sum()
```

```python
df.hist(bins=50, figsize=(20,15))
plt.savefig("Immunotherapy_Hist.png")
plt.show();
```

```python
import seaborn as sns

plt.figure(dpi = 120,figsize= (5,4))
mask = np.triu(np.ones_like(df.corr(),dtype = bool))
sns.heatmap(df.corr(),mask = mask, fmt = ".2f",annot=True,lw=1,cmap = 'plasma')
plt.yticks(rotation = 0)
plt.xticks(rotation = 90)
plt.title('Correlation Heatmap')
plt.savefig("Immunotherapy_Correlation.png")
plt.show()
```
